#Matplotlib practice 

- **Line**: best used when the x axis is time
- **Scatter** : best used when there's a correlation between variables
- **Histogram** : best used for data distribution

```python
 # Line Plot
data.Speed.plot(kind = 'line', color = 'g', label = 'Speed', linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
data.Defense.plot(color = 'r', label = 'Defense', linewidth =1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc = 'upper right') # puts the legend inside the plot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
```

```python
#Scatter Plot
# x = attack, y = defense
data.plot( kind = 'scatter', x = 'Attack', y ='Defense', alpha = 0.5, color = 'red')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack Defense Scatter Plot')
```

```python
# Histogram 
# bins = number of bar in figure
data.Speed.plot(kind = 'hist', color = 'black',  bins = 50, figsize = (12,12))
plt.show()
```

```python
 # clf() = cleans it up again so we can start anew
 data.Speed.plot(kind ='hist', color = 'r', bins = 50)
 plt.clf()
 #WE cannot see plot due to clf 
```

#Dictionary 

The reason we need it: 

- it is faster than lists
- it has a key and a value--> {'Spain': 'Madrid'}, where Spain is the key and Madrid si the value 

dictionary = {'Spain': 'Madrid', 'USA' : 'Vegas'} 



```python
# create dictionary and check the key and the value
dictionary = {'spain': 'madrid', 'usa': 'vegas'}
print(dictionary.keys())
print(dictionary.values())
```

```python
# Keys have to be immutable objects like strings, boolean, float, integers or tuples 
# Lists are not immutable, An immutable object is an object whose value cannot change.
# Keys are unique 
```

```python
dictionary['spain'] = 'barcelona'  #update existing entry 
print(dictionary)
dictionary['france'] = 'paris'    # add new entry 
print(dictionary)
del dictionary['spain']           # remove entry with key 'spain'
print('france' in dictionary)     # check if it includes or not 
dictionary.clear()                # remove all entries in dict
print(dictionary)
```

```python
#1 Filtering Pandas data frame 
x = data['Defense'] > 200 # Checking which pokemons have a Defense value higher than 200 
data[x]
```

```python
#2 Filtering pandas with logical_and 
# Checking which pokemons have a > defense value than 200 and > attack value than 100

data[np.logical_and(data['Defense'] > 200, data['Attack'] > 100)]
```

```python
data[(data['Defense'] > 200) & (data['Attack'] > 100)] # using the and operator to perform the same as above
```
